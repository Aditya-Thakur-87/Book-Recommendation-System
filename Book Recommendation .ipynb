# Popularity Based Recommendation System

```python
users_ratings = ratings.merge(books,on='ISBN')
```

```python
users_ratings.head()
```

```python
rating_df = users_ratings.groupby("Book-Title").count()['Book-Rating'].reset_index()
```

```python
rating_df
```

```python
rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
rating_df
```

```python
avg_rating_df = users_ratings.groupby("Book-Title")['Book-Rating'].mean().reset_index()
```

```python
users_ratings.info()
```

```python
avg_rating_df
```

```python
avg_rating_df.rename(columns={"Book-Rating": "Avg-Rating"}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df.head()
```

```python
popular_df=popular_df[popular_df['num_ratings']>250].sort_values('Avg-Rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author'
                                                                       ,'Image-URL-M','num_ratings','Avg-Rating']]
```

# Collaborating Filtering

```python
filt = users_ratings.groupby('User-ID').count()['Book-Rating']>200
```

```python
best_users = filt[filt].index
best_users
```

```python
filtered_rating = users_ratings[users_ratings['User-ID'].isin(best_users)]
```

```python
filt2 = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = filt2[filt2].index
```

```python
final_df = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt_df = final_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt_df.fillna(0,inplace=True)
pt_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_scores =cosine_similarity(pt_df)
```

```python
def recommend_books(book_name):
    index= np.where(pt_df.index==book_name)[0][0]
    print(index)
    similar_books = sorted(list(enumerate(sim_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    for index in similar_books:
        print(pt_df.index[index[0]])
```

```python
recommend_books('Nightmares &amp; Dreamscapes')
```
